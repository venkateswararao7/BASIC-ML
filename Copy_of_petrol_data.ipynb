```python
#import the  libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
print("set up is completed")
```

```python
y=my_data.Delhi
print(y)
print("completed")
```

```python
my_data_feature = ['Mumbai','Kolkata','Chennai']
x=my_data[my_data_feature]
print(x)
print("completed")
```

```python
from sklearn.tree import DecisionTreeRegressor 
data_model=DecisionTreeRegressor(random_state=1)
data_model.fit(x,y)
print("completed")
```

```python
print("the predicted value of 5 days:")
print(x.head())
print("the predictions are :")
print(data_model.predict(x.head()))
print(y.head())
print("completed")
```

```python
print(data_model.predict(x))
```

```python
print(y)
```

```python
from sklearn.metrics import mean_absolute_error
predicted_petrol_prize=data_model.predict(x)
print("the mean_absolute_error = ",end="")
print(mean_absolute_error(y,predicted_petrol_prize))
print('completed')
```

**Randomforestregressor**

```python
from sklearn.model_selection import train_test_split
train_x,val_x,train_y,val_y=train_test_split(x,y,random_state=0)
print("completed")
```

```python
train_x.head()
```

```python
from sklearn.ensemble import RandomForestRegressor
forest_model = RandomForestRegressor(random_state =1)
forest_model.fit(train_x,train_y)
my_data = forest_model.predict(val_x)
print(my_data)
print("the mean absolute error = ",end="")
print(mean_absolute_error(val_y,my_data))
print("completed")
```

**using XGBRegressor**


```python
from xgboost import XGBRegressor

my_model = XGBRegressor()
my_model.fit(train_x,train_y)
print("train of model is completed")
```

```python
my_data = my_model.predict(val_x)
print("the mean absolute error",end=" ")
print(mean_absolute_error(val_y,my_data))
print("training is completed")
```

from the above codes  we can understanded that the mean_absolute_error of using of  decisiontreeregressor  the mean_absolute_error = 0.0026292335115864745,
from Randomforestregressor the  mean absolute error = 0.1567204787233947,
from XGBregressor  the mean absolute error 0.19451010359094523.

from the three algorthm  decisiontreeregressor is working more efficiently when compare to other algorthm
